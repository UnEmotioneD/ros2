# Multiple Variable

````python
# Declare multiple variables in one line
n1, n2 = 10, 3


# Declare function or method
def calc(n1, n2):
    return n1 + n2, n1 - n2


result = calc(n1, n2)
print(result)  # type: tuple

# return multiple values
plus, minus = calc(n1, n2)
print(plus, minus)
````

### Set


````python
# Set: unordered, mutable, no duplicates

set1 = {1, 2, 3, 3, 3, 2}
set2 = {2, 3, 4, 5, 2}
set3 = set()

# there is no duplicates
print(f"set1: {set1}")
print(f"set2: {set2}")

print(f"set1 & set2: {set1 & set2}")
print(f"set1 | set2: {set1 | set2}")
print(f"set1 - set2: {set1 - set2}")
print(f"set2 - set1: {set2 - set1}")
````

### Dictionary

- no order
- mutable
- key must not be duplicated

```python
dictionary = {'key': value, 'foo': bar}
```


````python
dict1 = {"name": "hong", "age": 30, "addr": "seoul"}

# access with key
print(dict1["name"])
print(dict1["age"])
print(dict1["addr"])

# or like this
name, age, addr = dict1["name"], dict1["age"], dict1["addr"]
print(name, age, addr)
````

````python
dict2 = {0: "hong", 1: "lee", 2: "seo"}
print(dict2[0])  # use key to access

lst = ["hong", "lee", "seo"]
print(lst[0])  # use index to access

# both str class btw
````

````python
# Q: words starts with f

word_list = {
    "fubar": "It's a German.",
    "fake": "not genuine",
    "finger": "those things on your hands",
}

print(f'fubar: {word_list["fubar"]}')
print(f'fake: {word_list["fake"]}')
print(f'finger: {word_list["finger"]}')

word_list["fubar"] = "fucked up beyond any recognition"
print(f'fubar: {word_list["fubar"]}')
````
