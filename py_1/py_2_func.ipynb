````python
# isinstance()
def print_list_value(list) -> str:
    if isinstance(list, dict):
        for item in list:
            print(f"{item}: {list[item]}")
    else:
        for item in list:
            print(item)

    # without return it will iterate once more and return "None"
    return ""


def plv(list):
    if type(list) is dict:
        for item in list:
            print(f"{item}: {list[item]}")
    else:
        for item in list:
            print(item)


names = {"first": "kim", "middle": "yay", "last": "chan hee"}

print(print_list_value(names))
print(plv(names))
````

````python
# type check list, tuple, dic, set

lst = [1, 2, 3]
tpl = (1, 2, 3)
dic = {"a": 0, "b": 1, "c": 2}
st = {0, 1, 2}

print(type(lst))  # <class 'list'>
print(type(tpl))  # <class 'tuple'>
print(type(dic))  # <class 'dict'>
print(type(st))  # <class 'set'>
````

````python
def print_loop(args):
    if type(args) is int or type(args) is float:
        print("Can't iterate numbers.")
        return None
    elif type(args) is dict:
        for key in args:
            print(key, args[key])
    else:
        for item in args:
            print(item)

    return "All good"


# the_thing = 'hello'
the_thing = 1234

print_loop(the_thing)
````
